```python
import numpy as np
import pandas as pd
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"

data_path = '/Users/vikash/Downloads/DS TakeHome Assignment/data'
```

```python
%%time
pings = pd.read_csv(data_path + '/' + 'pings.csv')
from datetime import datetime, timedelta
pings['datetime_local'] = pings['ping_timestamp'].apply(lambda x: (datetime.fromtimestamp(x)).strftime('%Y-%m-%d %H:%M:%S'))
```

```python
%%time
pings['datetime_jkt'] = pings['ping_timestamp'].apply(lambda x: (datetime.utcfromtimestamp(x) + timedelta(hours=7)).strftime('%Y-%m-%d %H:%M:%S'))
```

```python
pings.head()
```

```python
%%time
pings['date_jkt'] = pings['datetime_jkt'].apply(lambda x: x.split(' ')[0])
pings['date_local'] = pings['datetime_local'].apply(lambda x: x.split(' ')[0])
```

### Grouping by Driver ID and Date

```python
%%time
df_pings = pings.groupby(['driver_id', 'date_local']).datetime_jkt.agg([('Min' , 'min'), ('Max', 'max')])\
       .add_prefix('Ping_')
```

```python
df_pings.reset_index(inplace=True)
df_pings.to_parquet('df_pings_local.parquet', engine='fastparquet', compression ='GZIP')
df_pings.head()
df_pings.shape
```
